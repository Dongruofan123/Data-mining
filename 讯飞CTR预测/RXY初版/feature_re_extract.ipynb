```python
import numpy as np
import pandas as pd
```

```python
#读取上一步中获得的数据
test_data = pd.read_csv('test_data.csv')
train_data = pd.read_csv('train_data.csv')
```

```python
#advert_industry_inner 字段中 有两类ID
def dataInterval(data1):
    d1 = data1.split('_')[0]
    return d1

def getInterval(arrLike):  #用来计算日期间隔天数的调用的函数
    PublishedTime = arrLike['advert_industry_inner']
#    print(PublishedTime.strip(),ReceivedTime.strip())
    d1 = dataInterval(PublishedTime)  #注意去掉两端空白
    return d1

def dataInterval2(data1):
    d2 = data1.split('_')[1]
    return d2

def getInterval2(arrLike):  #用来计算日期间隔天数的调用的函数
    PublishedTime = arrLike['advert_industry_inner']
#    print(PublishedTime.strip(),ReceivedTime.strip())
    d2 = dataInterval2(PublishedTime)  #注意去掉两端空白
    return d2
 
#使用apply要快速很多！
test_data['advert_first'] = test_data.apply(getInterval , axis = 1)
test_data['advert_second'] = test_data.apply(getInterval2 , axis = 1)

train_data['advert_first'] = train_data.apply(getInterval , axis = 1)
train_data['advert_second'] = train_data.apply(getInterval2 , axis = 1)
```

```python
#顺带计算两类ID间的差距
def getDistance(arrLike):
    delta = int(arrLike['advert_second']) - int(arrLike['advert_first'])
    return delta

test_data['advert_delta'] = test_data.apply(getDistance, axis = 1)
train_data['advert_delta'] = train_data.apply(getDistance, axis = 1)
```

```python
#删除原有advert_industry_inner字段
test_data.drop(columns=['advert_industry_inner'],  inplace=True)
train_data.drop(columns=['advert_industry_inner'],  inplace=True)
```

```python
#将所有类型转换为可以训练的
test_data['advert_first'] = test_data['advert_first'].astype('int64')
train_data['advert_first'] = train_data['advert_first'].astype('int64')

test_data['advert_second'] = test_data['advert_first'].astype('int64')
train_data['advert_second'] = train_data['advert_first'].astype('int64')
```

```python
#将click这个label移到最后
train_data = train_data[
    ['instance_id', 'time', 'city', 'province', 'user_tags', 'carrier',
       'devtype', 'nnt', 'os', 'os_name', 'adid',
       'advert_id', 'orderid', 'campaign_id',
       'creative_id', 'creative_tp_dnf', 'app_cate_id', 'f_channel', 'app_id',
       'creative_type', 'creative_width', 'creative_height',
       'creative_is_jump', 'creative_is_download', 'creative_is_js',
       'creative_is_voicead', 'creative_has_deeplink', 'app_paid','advert_first', 'advert_second', 'advert_delta',
       'click']
]
```
