```python
import xgboost as xgb
```

```python
model = xgb.Booster(model_file='model/xgb.model')
```

```python
%matplotlib inline
ax = xgb.plot_importance(model)
```

```python
import graphviz
```

```python
%matplotlib inline
bx = xgb.to_graphviz(model)
```

```python
cx = xgb.plot_tree(model, num_trees=0, rankdir='LR')
```

```python
bx = xgb.plot_tree(model)
```

```python
将决策树存成图片
import codecs  
f=codecs.open('xgb_tree.png', mode='wb')  
g=xgb.to_graphviz(model)  
f.write(g.pipe('png'));  
f.close()
```
